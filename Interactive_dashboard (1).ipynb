## Xử lý dữ liệu

```python
# Gán giá trị null bằng 0 và tạo cột GDP bình quân đầu người
df = df.fillna(0)
df['gdp_per_capita'] = np.where(df['population']!= 0, df['gdp']/ df['population'], 0)
```

## (1) CO2 emission over time by continent

```python
# Define Panel widgets
year_slider = pn.widgets.IntSlider(name='Year slider', start=1750, end=2020, step=1, value=2019)
year_slider
```

```python
select = pn.widgets.Select(name='Select', options=['co2', 'co2_per_capita','co2_per_gdp','co2_growth_abs','coal_co2_per_capita','co2_per_unit_energy','cement_co2_per_capita','coal_co2_per_capita','flaring_co2_per_capita','gas_co2_per_capita','oil_co2_per_capita','nitrous_oxide_per_capita','other_co2_per_capita'])

select
```

```python
continents = ['World', 'Asia', 'Oceania', 'Europe', 'Africa', 'North America', 'South America', 'Antarctica']

co2_pipeline = (
    idf[
        (idf.year <= year_slider) &
        (idf.country.isin(continents))
    ]
    .groupby(['country', 'year'])[select].mean()
    .to_frame()
    .reset_index()
    .sort_values(by='year')  
    .reset_index(drop=True)
)
co2_plot = co2_pipeline.hvplot(x = 'year', by='country', y=select,line_width=2, title="Lượng khí thải C02 hằng năm theo khu vực")
co2_plot
```

```python
select_1 = pn.widgets.Select(name='Select', options=['co2', 'co2_per_capita','co2_per_gdp','co2_growth_abs','coal_co2_per_capita','co2_per_unit_energy','cement_co2_per_capita','coal_co2_per_capita','flaring_co2_per_capita','gas_co2_per_capita','oil_co2_per_capita','nitrous_oxide_per_capita','other_co2_per_capita'])

select_1
```

```python
continents_1 = ['China', 'United States', 'India', 'Germany', 'Brazil', 'United Kingdom', 'France','Canada','Australia','Spain','Sweden']

co2_pipeline_1 = (
    idf[
        (idf.year <= year_slider) &
        (idf.country.isin(continents_1))
    ]
    .groupby(['country', 'year'])[select_1].mean()
    .to_frame()
    .reset_index()
    .sort_values(by='year')  
    .reset_index(drop=True)
)
co2_plot_1 = co2_pipeline_1.hvplot(x = 'year', by='country', y=select_1,line_width=2, title="Lượng khí thải C02 hằng năm theo quốc gia")
co2_plot_1
```

## (2) Bảng giá trị lượng khí thải C02

```python
#Bảng giá trị lượng khí thải theo khu vực
co2_table = co2_pipeline.pipe(pn.widgets.Tabulator, pagination='remote', page_size = 12, sizing_mode='stretch_width') 
co2_table
```

```python
# Bảng giá trị lượng khí thải theo quốc gia
co2_table_1 = co2_pipeline_1.pipe(pn.widgets.Tabulator, pagination='remote', page_size = 12, sizing_mode='stretch_width') 
co2_table_1
```


## (3)Biểu đồ scatterplot CO2 dựa trên sản xuất

```python
select_2 = pn.widgets.Select(name='Select', options=['coal_co2','gas_co2','oil_co2','cement_co2','flaring_co2','methane','nitrous_oxide','gdp_per_capita','co2_per_capita','co2_growth_abs'])
```

```python
co2_vs_gdp_scatterplot_pipeline = (
    idf[
        (idf.year == year_slider) &
        (~ (idf.country.isin(continents)))
    ]
    .groupby(['country', 'year','co2'])[select_2].mean()
    .to_frame()
    .reset_index()
    .sort_values(by='year')  
    .reset_index(drop=True)
)
co2_vs_gdp_scatterplot = co2_vs_gdp_scatterplot_pipeline.hvplot(x='co2', 
                                                                y=select_2,
                                                                by='country', 
                                                                size=80, kind="scatter", 
                                                                alpha=0.7,
                                                                legend=False, 
                                                                height=500, 
                                                                width=500,
                                                                title=('Tỷ lệ phát thải CO2 dựa trên sản xuất'))
co2_vs_gdp_scatterplot
```

## (4)Biểu đồ scatterplot CO2 dựa trên mức tiêu thụ

```python
select_3 = pn.widgets.Select(name='Select', options=['consumption_co2','consumption_co2_per_capita','consumption_co2_per_gdp','energy_per_capita','energy_per_gdp'])

co2_vs_gdp_scatterplot_pipeline_1 = (
    idf[
        (idf.year == year_slider) &
        (~ (idf.country.isin(continents)))
    ]
    .groupby(['country', 'year','co2'])[select_3].mean()
    .to_frame()
    .reset_index()
    .sort_values(by='year')  
    .reset_index(drop=True)
)
co2_vs_gdp_scatterplot_1 = co2_vs_gdp_scatterplot_pipeline_1.hvplot(x='co2', 
                                                                y=select_3,
                                                                by='country', 
                                                                size=80, kind="scatter", 
                                                                alpha=0.7,
                                                                legend=False, 
                                                                height=500, 
                                                                width=500,
                                                                title=('Lượng khí thải CO2 dựa trên mức tiêu thụ'))
co2_vs_gdp_scatterplot_1
```

## (5)Biểu đồ bar chart các loại khí thải theo lục địa

```python
yaxis_co2_source = pn.widgets.RadioButtonGroup(
    name='Y axis', 
    options=['coal_co2', 'oil_co2', 'gas_co2','cement_co2','flaring_co2','methane','nitrous_oxide'], 
    button_type='success'
)

continents_excl_world = ['Asia', 'Oceania', 'Europe', 'Africa', 'North America', 'South America', 'Antarctica']

co2_source_bar_pipeline = (
    idf[
        (idf.year == year_slider) &
        (idf.country.isin(continents_excl_world))
    ]
    .groupby(['year', 'country'])[yaxis_co2_source].sum()
    .to_frame()
    .reset_index()
    .sort_values(by='year')  
    .reset_index(drop=True)
)
```

```python
co2_source_bar_plot = co2_source_bar_pipeline.hvplot(kind='bar', 
                                                     x='country', 
                                                     y=yaxis_co2_source, 
                                                     title='Các loại khí thải gây hiệu ứng nhà kính giữa các khu vực')
co2_source_bar_plot
```

## (6) Scatterplot CO2 và GDP đầu người 

```python
co2_vs_gdp_scatterplot_pipeline_2 = (
    idf[
        (idf.year == year_slider) &
        (~ (idf.country.isin(continents)))
    ]
    .groupby(['country', 'year', 'gdp_per_capita'])['co2'].mean()
    .to_frame()
    .reset_index()
    .sort_values(by='year')  
    .reset_index(drop=True)
)
```

```python
co2_vs_gdp_scatterplot_2 = co2_vs_gdp_scatterplot_pipeline_2.hvplot(x='gdp_per_capita', 
                                                                y='co2', 
                                                                by='country', 
                                                                size=80, kind="scatter", 
                                                                alpha=0.7,
                                                                legend=False, 
                                                                height=500, 
                                                                width=500,
                                                                title=('Phát thải CO2 trên một đơn vị GDP'))
co2_vs_gdp_scatterplot_2
```
